```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
```

```python
train = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/Datarisk/data/processed_train.csv')
```

## 3.1 Quartis e outliers

```python
for i in train:
  ax = sns.boxplot(x='INADIMPLENTE', y=i, data=train)
  plt.show()
```

## 3.2 Distribuição de valores

```python
for i in train:
  ax = sns.histplot(data=train, x=i, hue ='INADIMPLENTE')
  plt.xticks(rotation=90)
  plt.show()
```

## 3.3 Médias detalhadas

```python
binsAC = pd.cut(train['ANO_CADASTRO'], 10)
display(train.groupby(binsAC).mean().sort_values(by='INADIMPLENTE', ascending=False))
```

```python
binsNF = pd.cut(train['NO_FUNCIONARIOS'], 10)
display(train.groupby(binsNF).mean().sort_values(by='INADIMPLENTE', ascending=False))
```

```python
binsRMA = pd.cut(train['RENDA_MES_ANTERIOR'], 10)
display(train.groupby(binsRMA).mean().sort_values(by='INADIMPLENTE', ascending=False))
```
